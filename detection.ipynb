```python
import os

from PIL import Image
from IPython.display import display
from roboflow import Roboflow
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv('ROBOFLOW_API_KEY')
```

```python
rf = Roboflow(api_key=API_KEY)
project = rf.workspace().project("banana-ripening-process")
model = project.version(2).model

# infer on a local image & visualize your prediction
model.predict("banana_1.jpg", confidence=40, overlap=50).save("output.jpg")
# print(model.predict("banana_2.jpg", confidence=30, overlap=30).json())

# infer on an image hosted elsewhere
# print(model.predict("URL_OF_YOUR_IMAGE", hosted=True, confidence=40, overlap=30).json())
```
